```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import accuracy_score
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.info()
```

```python
print(df["sex"].value_counts())
print(df["day"].value_counts())
print(df["smoker"].value_counts())
print(df["time"].value_counts())
```

```python
X=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
X_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lab1=LabelEncoder()
lab2=LabelEncoder()
lab3=LabelEncoder()
```

```python
X_train["smoker"]=lab1.fit_transform(X_train["smoker"])
X_train["sex"]=lab2.fit_transform(X_train["sex"])
X_train["time"]=lab3.fit_transform(X_train["time"])
```

```python
X_test["smoker"]=lab1.transform(X_test["smoker"])
X_test["sex"]=lab2.transform(X_test["sex"])
X_test["time"]=lab3.transform(X_test["time"])
```

```python
X_test
```

```python
X_test
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct=ColumnTransformer(transformers=[("onehot",OneHotEncoder(drop="first"),[3])],remainder="passthrough")
```

```python
ct.fit_transform(X_train)
```

```python
import sys, numpy
numpy.set_printoptions(threshold=sys.maxsize)
X_train=ct.fit_transform(X_train)
```

```python
X_test=ct.transform(X_test)
```

```python
X_test
```

```python
Svr=SVR()
Svr.fit(X_train,y_train)
```

```python
y_predict=Svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
r2=r2_score(y_test,y_predict)
mae=mean_absolute_error(y_test,y_predict)
```

```python
r2,mae
```

```python
from sklearn.model_selection import GridSearchCV

param_grid={
    "C":[0.1,1,10,100,1000],
    "gamma":[1,0.1,0.01,0.001,0.0001],
    "kernel":["rbf"]
}
```

```python
grid=GridSearchCV(estimator=SVR(),param_grid=param_grid,refit=True,verbose=3)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
r2=r2_score(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
```
