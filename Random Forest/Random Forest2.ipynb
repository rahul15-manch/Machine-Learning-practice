```python
df["Gender"]=df["Gender"].replace({"Fe Male":"Female"})
df["Gender"].value_counts()
```

```python
feature_with_na=[feature for feature in df.columns if df[feature].isnull().sum()>=1]
for feature in feature_with_na:
    print(feature,np.round(df[feature].isnull().mean()*100,5),"% missing value")
```

```python
df["Age"].fillna(df["Age"].median(),inplace=True)
```

```python
df["TypeofContact"].fillna(df["TypeofContact"].mode()[0],inplace=True)
df["DurationOfPitch"].fillna(df["DurationOfPitch"].median(),inplace=True)
df["TypeofContact"].fillna(df["TypeofContact"].mode()[0],inplace=True)
df["NumberOfFollowups"].fillna(df["NumberOfFollowups"].mode()[0],inplace=True)
df.PreferredPropertyStar.fillna(df.PreferredPropertyStar.mode([0]),inplace=True)
df.NumberOfTrips.fillna(0,inplace=True)
df["NumberOfChildrenVisiting"].fillna(df["NumberOfChildrenVisiting"].mode()[0],inplace=True)
```

```python
df["MonthlyIncome"].fillna(df["MonthlyIncome"].median(),inplace=True)
df["PreferredPropertyStar"].fillna(df["PreferredPropertyStar"].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
df["NumberOfTrips"]=df["NumberOfTrips"].replace({0:np.nan})
```

```python
df.NumberOfTrips.fillna(df.NumberOfTrips.median(),inplace=True)
```

```python
df.drop("CustomerID", inplace=True, axis=1)
```

```python
df["TotalVisting"]=df["NumberOfChildrenVisiting"]+df["NumberOfPersonVisiting"]
```

```python
df1=pd.read_csv("Travel.csv")
df["ProdTaken"]=df1["ProdTaken"]
```

```python
X=df
y=df1["ProdTaken"]
```

```python
X=X.drop("ProdTaken",axis=1)
```

```python
y.value_counts()
```

```python
X.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
X_train.head()
```

```python
cat_feature=X.select_dtypes(include="object").columns
num_feature=X.select_dtypes(exclude="object").columns
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
scale=StandardScaler()
encode=OneHotEncoder(drop="first")
ct=ColumnTransformer([
        ("OneHotEncoder",encode,cat_feature),
        ("StandardScaler",scale,num_feature)
    ])
```

```python
X_train=ct.fit_transform(X_train)
X_test=ct.transform(X_test)
```

```python
ct
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix,precision_score,recall_score,f1_score,roc_auc_score,roc_curve,ConfusionMatrixDisplay
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier


models = {
    "Random Forest": RandomForestClassifier()
    ,"Decision Tree":DecisionTreeClassifier()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    
    # Predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Training Metrics
    model_train_accuracy = accuracy_score(y_train, y_train_pred)
    model_train_f1 = f1_score(y_train, y_train_pred)
    model_train_precision = precision_score(y_train, y_train_pred)
    model_train_recallscore = recall_score(y_train, y_train_pred)
    model_train_roc = roc_auc_score(y_train, y_train_pred)

    # Testing Metrics
    model_test_accuracy = accuracy_score(y_test, y_test_pred)
    model_test_f1 = f1_score(y_test, y_test_pred)
    model_test_precision = precision_score(y_test, y_test_pred)
    model_test_recallscore = recall_score(y_test, y_test_pred)
    model_test_roc = roc_auc_score(y_test, y_test_pred)

    print(f"\nModel: {name}")
    print(f"Train Accuracy: {model_train_accuracy:.4f}, F1: {model_train_f1:.4f}, Precision: {model_train_precision:.4f}, Recall: {model_train_recallscore:.4f}, ROC-AUC: {model_train_roc:.4f}")
    print(f"Test  Accuracy: {model_test_accuracy:.4f}, F1: {model_test_f1:.4f}, Precision: {model_test_precision:.4f}, Recall: {model_test_recallscore:.4f}, ROC-AUC: {model_test_roc:.4f}")
```

```python
print(model_train_accuracy,model_test_accuracy)
print(model_train_f1,model_test_f1)
print(model_train_recallscore,model_test_recallscore)
print(model_train_roc,model_test_roc)
print(model_train_precision,model_test_precision)
```
